```python
from IPython import display
def use_svg_display():  #@save
    """使用svg格式在Jupyter中显示绘图"""
    display.set_matplotlib_formats('svg')
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
feature_importance = pd.read_hdf("feature_importance.hdf")
```

```python
best_features = feature_importance.iloc[:20].reset_index()
best_features = best_features[['feature', 'importance_split']]
best_features.rename({'importance_split': 'importance'}, axis=1, inplace=True)

import matplotlib.gridspec as gridspec
fig = plt.figure(figsize=(8, 8), dpi = 300)
gs = gridspec.GridSpec(1, 1)
# Plot Split importances
ax = plt.subplot(gs[0, 0])
sns.barplot(x='importance', y='feature', data=best_features, ax=ax)
ax.set_title('Feature importance', fontweight='bold', fontsize=14)
```
